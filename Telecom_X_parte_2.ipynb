```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
X = df.drop('Cliente_cancelado', axis=1)
y = df['Cliente_cancelado']
columnas = X.columns
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary'),['Tiene_Dependientes','Mayor_de_65_años','Tipo_de_Contrato','Metodo_Pago']), remainder='passthrough',sparse_threshold=0,force_int_remainder_cols=False)
X = one_hot.fit_transform(X)
one_hot.get_feature_names_out(columnas)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X,X_test,y, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state= 5)
X_train,X_val,y_train, y_val   = train_test_split(X, y, test_size=0.15, stratify=y, random_state= 5)
Baseline = DecisionTreeClassifier(max_depth=10)
Baseline.fit(X_train, y_train)
Baseline.score(X_val, y_val)
```

```python
from sklearn.metrics import classification_report
y_pred = Baseline.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
from sklearn.model_selection import KFold,cross_validate
kf = KFold(n_splits = 5, shuffle = True, random_state=5)
cv_resultados = cross_validate (Baseline,X,y,cv=kf)
promedio = cv_resultados['test_score'].mean()
desvio_std = cv_resultados['test_score'].std()
print(f'El promedio es: {promedio}')
print(f'El desvio estandar es: {desvio_std}')
print(f'El intervalo de confianza es: [{promedio - 2*desvio_std}, {min(promedio +2*desvio_std,1)}')
```

```python
def intervalo_confianza (resultados):
  promedio = resultados['test_score'].mean()
  desvio_std = resultados['test_score'].std()
  return f'El intervalo de confianza es: [{promedio - 2*desvio_std}, {min(promedio +2*desvio_std,1)}'
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state=5)
cv_resultados = cross_validate (Baseline,X,y,cv=skf, scoring='recall')
intervalo_confianza(cv_resultados)
```

Balanceo de las Clases OverSampling

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

X = df.drop('Cliente_cancelado', axis=1)
y = df['Cliente_cancelado']
columnas = X.columns
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary'),['Tiene_Dependientes','Mayor_de_65_años','Tipo_de_Contrato','Metodo_Pago']), remainder='passthrough',sparse_threshold=0,force_int_remainder_cols=False)
X = one_hot.fit_transform(X)
one_hot.get_feature_names_out(columnas)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd
over_sampling = SMOTE()
X_balanceada_over,y_balanceada_over = over_sampling.fit_resample(X,y)
pd.Series(y_balanceada_over).value_counts()
```

```python
from sklearn.model_selection import StratifiedKFold
modelo = DecisionTreeClassifier(max_depth=10)
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state=5)
cv_resultados = cross_validate (modelo,X_balanceada_over,y_balanceada_over,cv=skf, scoring='recall')
intervalo_confianza(cv_resultados)
```

Validación

**Con los datos balanceados y para validar, debemos hacerlo con los datos reales**

```python
from imblearn.pipeline import Pipeline as imbpipeline
modelo = DecisionTreeClassifier(max_depth=10)
pipeline = imbpipeline([('oversample', SMOTE()), ('arbol', modelo)])
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state=5)
cv_resultados = cross_validate (pipeline,X,y,cv=skf, scoring='recall')
intervalo_confianza(cv_resultados)
```

Paso 7: Separaremos los datos

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import NearMiss
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

url = "https://raw.githubusercontent.com/Deinbi/Telecom-X-Parte-2-Predicci-n-de-Cancelaci-n-Churn-/refs/heads/main/datos_tratados.csv"
df = pd.read_csv(url)

X = df.drop('Cliente_cancelado', axis=1)
y = df['Cliente_cancelado']

# Drop the 'ID_Cliente' column before splitting
X = X.drop('ID_Cliente', axis=1)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, stratify=y, random_state=5)

# Apply one-hot encoding to categorical columns in the training data
categorical_cols = X_train.select_dtypes(include='object').columns
one_hot = make_column_transformer(
    (OneHotEncoder(drop='first'), categorical_cols),
    remainder='passthrough',
    sparse_threshold=0,
    force_int_remainder_cols=False
)

X_train_encoded = one_hot.fit_transform(X_train)

# Apply NearMiss to the encoded training data
undersample = NearMiss(version=3)
X_balanceado, y_balanceado = undersample.fit_resample(X_train_encoded, y_train)
```

Paso 8: Creamos modelos

Modelo de árbol de decisiones

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

modelo = DecisionTreeClassifier(max_depth=10)
modelo.fit(X_balanceado, y_balanceado)

# Before predicting on X_val, it needs to be transformed using the same one_hot encoder fitted on X_train
X_val_encoded = one_hot.transform(X_val)

y_previsto = modelo.predict(X_val_encoded)

print(classification_report(y_val, y_previsto))

ConfusionMatrixDisplay.from_predictions(y_val, y_previsto);
```

Modelo KNN

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import NearMiss
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier

url = "https://raw.githubusercontent.com/Deinbi/Telecom-X-Parte-2-Predicci-n-de-Cancelaci-n-Churn-/refs/heads/main/datos_tratados.csv"
df = pd.read_csv(url)

X = df.drop('Cliente_cancelado', axis=1)
y = df['Cliente_cancelado']

X = X.drop('ID_Cliente', axis=1)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, stratify=y, random_state=5)

categorical_cols = X_train.select_dtypes(include='object').columns
one_hot = make_column_transformer(
    (OneHotEncoder(drop='first'), categorical_cols),
    remainder='passthrough',
    sparse_threshold=0,
    force_int_remainder_cols=False
)

X_train_encoded = one_hot.fit_transform(X_train)
X_val_encoded = one_hot.transform(X_val)

undersample = NearMiss(version=3)
X_balanceado, y_balanceado = undersample.fit_resample(X_train_encoded, y_train)

modelo_knn = KNeighborsClassifier(n_neighbors=5)
modelo_knn.fit(X_balanceado, y_balanceado)

y_pred_knn = modelo_knn.predict(X_val_encoded)

print("KNN (k=5):")
print(classification_report(y_val, y_pred_knn))

ConfusionMatrixDisplay.from_predictions(y_val, y_pred_knn);
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import NearMiss
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder


url = "https://raw.githubusercontent.com/Deinbi/Telecom-X-Parte-2-Predicci-n-de-Cancelaci-n-Churn-/refs/heads/main/datos_tratados.csv"
df = pd.read_csv(url)

X = df.drop('Cliente_cancelado', axis=1)
y = df['Cliente_cancelado']

X = X.drop('ID_Cliente', axis=1)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, stratify=y, random_state=5)

categorical_cols = X_train.select_dtypes(include='object').columns
one_hot = make_column_transformer(
    (OneHotEncoder(drop='first'), categorical_cols),
    remainder='passthrough',
    sparse_threshold=0,
    force_int_remainder_cols=False
)

X_train_encoded = one_hot.fit_transform(X_train)
X_val_encoded = one_hot.transform(X_val)

undersample = NearMiss(version=3)
X_balanceado, y_balanceado = undersample.fit_resample(X_train_encoded, y_train)


scaler_minmax = MinMaxScaler()
X_balanceado_norm = scaler_minmax.fit_transform(X_balanceado)
X_val_norm = scaler_minmax.transform(X_val_encoded)

modelo_knn_norm = KNeighborsClassifier(5)
modelo_knn_norm.fit(X_balanceado_norm, y_balanceado)

y_pred_knn_norm = modelo_knn_norm.predict(X_val_norm)

print("KNN (k=5) con Normalización Min-Max:")
print(classification_report(y_val, y_pred_knn_norm))

ConfusionMatrixDisplay.from_predictions(y_val, y_pred_knn_norm);
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.utils.validation import check_is_fitted

def debug_feature_importances(model=None, X=None):
    # Buscar variables comunes en el entorno global si no se pasan
    g = globals()
    if model is None:
        for name in ['modelo_rf','rf_model','model_rf','modelo_random_forest','rf']:
            if name in g:
                model = g[name]
                print(f"> Encontrado modelo en variable: {name}")
                break
    if model is None: # Check again if model is still None after searching
        print("ERROR: No se encontró ninguna variable de modelo. Define tu RandomForest entrenado en 'modelo_rf' o pásalo como parámetro.")
        return

    if X is None:
        for name in ['X_train_encoded','X','X_train_final','X_train_scaled']: # Look for the encoded training data
            if name in g:
                X = g[name]
                print(f"> Encontrado X en variable: {name}")
                break
    if X is None: # Check again if X is still None after searching
        print("ERROR: No se encontró ninguna variable X (ej. X_train_encoded). Definela o pásala como parámetro.")
        return


    # Si es Pipeline, extraer el estimador final
    if isinstance(model, Pipeline):
        print("> El objeto es un Pipeline — extrayendo estimador final.")
        final = model.steps[-1][1]
    else:
        final = model

    # Comprobar si el estimador está entrenado
    try:
        check_is_fitted(final)
    except Exception as e:
        print("ADVERTENCIA: El estimador final parece NO estar entrenado o check_is_fitted falló:", e)

    # Comprobar atributo feature_importances_
    if not hasattr(final, 'feature_importances_'):
        print("ERROR: El estimador final NO tiene 'feature_importances_'. Tipo de estimador:", type(final))
        print("¿Usaste un modelo que no es árbol (ej. LogisticRegression) o olvidaste entrenarlo?")
        return

    importances = final.feature_importances_

    # Attempt to get feature names from the one_hot encoder
    try:
        # Assuming one_hot is available in the global scope and fitted
        if 'one_hot' in g and hasattr(g['one_hot'], 'get_feature_names_out'):
             colnames = list(g['one_hot'].get_feature_names_out())
             print("> Nombres de características obtenidos del one_hot encoder.")
        else:
             print("ADVERTENCIA: No se pudo obtener nombres de características del one_hot encoder.")
             raise AttributeError("one_hot encoder not found or not fitted")
    except Exception:
        # Fallback if one_hot encoder is not available or doesn't have get_feature_names_out
        try:
            # if X is a DataFrame
            colnames = list(X.columns)
            print("> Nombres de características obtenidos del DataFrame X.")
        except Exception:
            # if X is an array
            colnames = [f"X{i}" for i in range(X.shape[1])]
            print("> Nombres de características generados (X0, X1, ...).")


    print(f"> n_features en X: {X.shape[1]}; longitud de importances: {len(importances)}")

    # If there's a mismatch in lengths, try to handle it
    if len(importances) != len(colnames):
        print("ADVERTENCIA: La longitud de feature_importances_ NO coincide con el número de columnas de X o nombres de características.")
        print("Intentando emparejar por el menor tamaño.")
        m = min(len(importances), len(colnames))
        df_imp = pd.DataFrame({'Variable': colnames[:m], 'Importancia': importances[:m]})
    else:
        df_imp = pd.DataFrame({'Variable': colnames, 'Importancia': importances})

    df_imp = df_imp.sort_values('Importancia', ascending=False)
    print("\nTop 10 variables según importancias:")
    print(df_imp.head(10).to_string(index=False))

    # Graficar top 10 if they exist
    top = df_imp.head(10)
    if not top.empty:
        plt.figure(figsize=(10,6))
        sns.barplot(data=top, x='Importancia', y='Variable')
        plt.title('Importancia de Variables - Random Forest')
        plt.tight_layout()
        plt.show()
    else:
        print("No hay variables para graficar.")

# Call the function with the Random Forest model
debug_feature_importances(model=modelo_rf, X=X_train_encoded)
```
